# Training a Neural Network

```python
# Loading the dataset
import numpy as np
X_full_enc = np.load('X_full_enc.npy')
Y_full = np.load('Y_full.npy')
X_eval_enc = np.load('X_eval_enc.npy')
```

```python
X_full_enc.shape, Y_full.shape, X_eval_enc.shape
```

```python
# Building the model using TensorFlow and Keras
import tensorflow as tf

n_classes = len(np.unique(Y_full))

learning_rate = 0.001
epochs = 225

# Preparing the model
nn_tensorflow = tf.keras.Sequential([
    # The hidden layers
    tf.keras.layers.Dense(32, activation = 'relu', input_shape = [X_full_enc
                          .shape[1],]),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    # The output layer
    tf.keras.layers.Dense(n_classes, activation = 'softmax')
])

# Compiling the model
nn_tensorflow.compile(optimizer = tf.keras.optimizers.Adam(learning_rate =
                                                           learning_rate),
                      loss='sparse_categorical_crossentropy', metrics='accuracy')


# Adding stop criteria: 
from tensorflow.keras.callbacks import EarlyStopping

# Define early stopping
early_stopping = EarlyStopping(monitor='loss', patience=20)


# Fitting the model
nn_tensorflow.fit(X_full_enc, Y_full, epochs = epochs, verbose = 1, callbacks=[early_stopping])
```

## Predicting on the evaluation dataset

```python
print("Generate predictions for test samples")
predictions = nn_tensorflow.predict(X_eval_enc)
```

```python
import json

# Using argmax to get the actual prediction (class)
predictions_classes = predictions.argmax(axis=-1)

# Convert predictions to int and then to list
predictions_list = predictions_classes.astype(int).tolist()

# Prepare the data for the JSON file
data = {"target": dict(enumerate(predictions_list))}

# Write to the JSON file
with open('predictions.json', 'w') as outfile:
    json.dump(data, outfile)
```
